## Import Libraries

````python
from pathlib import Path
import sys

project_root = Path.cwd().resolve().parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from src.utils import set_cover_config, config_to_dict
from src.generator import generate_blog_cover

from IPython.display import Image as DisplayImage, display
````

---

## Define Parameters

````python
# Title and subtitle for the blog cover.
title = "How to Build a Blog Cover Generator\nUsing Pillow in Python"
subtitle = "Written by A Muh Mufqi Zuhudi"
````

````python
# Configuration for the blog cover
config = set_cover_config(
    width=1000,
    height=420,
    background_color="#0f172a",
    text_color="#dedadc",
    font_size_title=45,
    font_size_subtitle=15,
    vertical_margin=15,
    line_gap=15,
    output_format="PNG",
    output_dir="./output",
    logo_size=30,
    logo_margin=15,
)
````

````python
# Provide local font file paths and logo path if you have custom files.
title_font_path = "../assets/fonts/raleway-thin.ttf"
subtitle_font_path = "../assets/fonts/raleway.ttf"

logo_path = "../assets/img/logo.png"
````

````python
# Convert config to dictionary
config_to_dict(config)
````

---

## Generate Blog Cover

````python
output_path = generate_blog_cover(
    title=title,
    subtitle=subtitle,
    config=config,
    title_font_path=title_font_path,
    subtitle_font_path=subtitle_font_path,
    logo_path=logo_path,
)

print(f"Generated: {output_path}")
display(DisplayImage(filename=str(output_path)))
````

## Optional: Using Custom Fonts and Logo

To use custom fonts, place `.ttf` font files in an `./assets/fonts/` directory:

```
project-root/
  assets/
    fonts/
      raleway.ttf
      raleway-thin.ttf
    img/
      logo.png
```

Then in the parameters cells above, update:
- `title_font_path = "../assets/fonts/raleway-thin.ttf"`
- `subtitle_font_path = "../assets/fonts/raleway.ttf"`
- `logo_path = "../assets/img/logo.png"`

If fonts or logo are not found, the generator will automatically use system fonts (like Arial on macOS) with the specified sizes, and skip the logo.
